```python
import os
import pandas as pd
import json
```

```python
def goodreads():  #goodreadsList as argument


    #os.system("scrapy crawl GoodReads -o c.json")

  
    file = open("d.json")

    data = json.load(file) #data is a list of dictionaries, where key=url, values = list of other versions name for the book

    dataframe = pd.DataFrame(data)
    
    #----------------
    #Clean Data: 
    #-------------------------------

    
    #join duplicate urls
    dataframe = dataframe.groupby(["url"]).agg('sum')
    
    #end of line things between parantheses like (Kindle Edition) or (Hardcover)
        #regex = (\([a-z-A-Z\s]*?\))$  -> check for every item in the list if it ends with this regex
                                          #if it does so, then split and return the part in the left of the parenthesis 
                                                                
    


    #-----------------------
    #-----------------------------------
    
    return dataframe
goodreads()
```
